# 5 - Inferencing

```python
# Split between num column and cat column
num_col_skew = ['limit_balance', 
                'bill_amt_1', 'bill_amt_2', 'bill_amt_3', 'bill_amt_4', 'bill_amt_5', 'bill_amt_6', 
                'pay_amt_1', 'pay_amt_2', 'pay_amt_3', 'pay_amt_4', 'pay_amt_5', 'pay_amt_6']
cat_col_ordinal = ['pay_1', 'pay_2', 'pay_3', 'pay_4', 'pay_5', 'pay_6']
cat_col_nominal = ['age_cat', 'sex', 'education_level', 'marital_status']

data_inf_num_s = df[num_col_skew]
data_inf_cat_o = df[cat_col_ordinal]
data_inf_cat_n = df[cat_col_nominal]
```

```python
data_inf_num_s_scaled = model_scaler_mm.transform(data_inf_num_s)
data_inf_cat_n_encoded = model_oh_encoder.transform(data_inf_cat_n).toarray()
```

```python
data_inf_final = np.concatenate([data_inf_num_s_scaled, data_inf_cat_o, data_inf_cat_n_encoded], axis=1)
```

```python
threshold = 0.6
y_pred_new_proba_train = model_knn.predict_proba(data_inf_final)
y_pred_new_train = np.where(y_pred_new_proba_train[:,1] >= threshold, 1, 0)
```
